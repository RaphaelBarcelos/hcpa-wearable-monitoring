```python
# --- 1. Carregamento de Dados REAIS do PAMAP2 ---

# 1.1 Configuração e Carregamento (Corrigido para a raiz do projeto)
nome_arquivo = 'datasets/dataset2.csv' 
# Se você tiver certeza que o separador é o ponto e vírgula (;), mude 'sep=',' para 'sep=';'
df = pd.read_csv(nome_arquivo, sep=',')

# 1.2 Limpeza dos Nomes das Colunas
# Remove o símbolo '#' e espaços em branco dos nomes das colunas (Ex: '# heart_rate' -> 'heart_rate')
df.columns = df.columns.str.replace('#', '').str.strip().str.lower()

# 1.3 Renomear colunas para o padrão do nosso modelo
# Assumimos que a coluna de Batimentos limpa é 'heart_rate' e a de Atividade é 'activityid'
df.rename(columns={'heart_rate': 'bpm', 
                   'activityid': 'target'}, 
          inplace=True)

# 1.4 Codificação do Rótulo (Label Encoding)
# Transforma as strings ('walking', 'lying', etc.) em números (0, 1, 2...)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Se o rótulo for 'transient activities' (atividades transitórias), remova antes de codificar,
# pois elas não têm valor preditivo.
df = df[df['target'] != 'transient activities'].copy()

# Sobrescreve a coluna 'target' com os valores numéricos
df['target'] = le.fit_transform(df['target']) 

# 1.5 Filtragem Essencial
# Remove linhas com valores ausentes (NaN)
df.dropna(inplace=True)


# Exibe o mapeamento das classes (para saber qual número representa qual atividade)
print("\n--- Mapeamento das Atividades (Rótulos Numéricos) ---")
print(dict(zip(le.classes_, le.transform(le.classes_))))

print("\n--- Carregamento Concluído com dados reais ---")
print(f"Total de linhas após limpeza e codificação: {len(df)}")
print("Colunas prontas para Feature Engineering:")
print(df[['bpm', 'target']].head())
```
